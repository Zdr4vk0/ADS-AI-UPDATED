```python
# Deze cel alleen uitvoeren, NIETS veranderen! 
import pandas as pd
import numpy as np
#dataframe:
df = pd.read_csv('WorldSustainabilityDataset.csv')
df2018 = df[df['Year'] == 2018]
inc_gini = df2018[['Country Name', 'Country Code', 
                   'GDP per capita (current US$) - NY.GDP.PCAP.CD', 
                   'Gini index (World Bank estimate) - SI.POV.GINI', 'Income Classification (World Bank Definition)']]
inc_gini.columns = ['Country Name', 'Country Code','GDP per capita', 'Gini index', 'Income class']
inc_gini = inc_gini.dropna()
inc_gini_df = inc_gini.set_index('Country Name')

#lists
country = list(inc_gini['Country Name'])
cc = list(inc_gini['Country Code'])
gdp = list(inc_gini['GDP per capita'])
gini = list(inc_gini['Gini index'])

#dictionary
inc_gini_d = {a:{'Country': b, 'GDP': c, 'Gini': d } for a, b, c, d in zip(cc, country, gdp, gini)}

#numpy array
np.set_printoptions(suppress=True)
inc_gini_n = inc_gini[['Country Code','GDP per capita', 'Gini index']].values

!conda --version
!jupyter notebook --version
!python --version
!anaconda-navigator --version
```

```python
#2a, Het inc gini index is in kolom 3 met kolom index 2
median = np.median(inc_gini_n[:, 2])
print(f' Het mediane gini index is: {median.round(2)}')
```

```python
#2b
npa = inc_gini_n
hoger_dan35 = npa[:, 2] > 35
hoofd5 = npa[:, 1]> 50000
selectie = npa[(hoger_dan35) & (hoofd5)]
print(f'Landen met een gini index hoger dan 35 en grotere gdp dan 50000:\n {selectie}')
```

```python
#2c
gdp = inc_gini_n[:, 1]
gemiddeld = inc_gini_n[:,1].mean()
groter_dan_gem = inc_gini_n[gdp>gemiddeld]
print(f'DE landcodes van de landen met een inkomen per hoofd (GDP) hoger dan gemiddeld zijn:\n \n{groter_dan_gem[:, 0]}')
```
