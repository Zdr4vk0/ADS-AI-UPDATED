```python
#Verander niets en voer onderstaande cel uit
import pandas as pd
import numpy as np
#dataframe:
df = pd.read_csv('storedata_clean.csv',sep = ';', index_col = 'Store ID')
df['Town'] = df['Town'].str.strip()
#lists:
town = list(df['Town'])
manager = list(df['Manager name'])
profit = list(df['Profit'])
#numpy array
st_fl_np = df[['Town', 'Floor Space']].values
#dictionary
st = {t:{'manager':m, 'profit':p} for t,m,p in zip(town,manager,profit)}


!conda --version
!jupyter notebook --version
!python --version
!anaconda-navigator --version
```

```python
#1c
print(max(profit))
index = profit.index(max(profit))
print(f'De town met het meeste winst is {town[index]}')
#1d
maxmanager = manager[index]
print(f'De manager met de meeste winst is {maxmanager}')
```
